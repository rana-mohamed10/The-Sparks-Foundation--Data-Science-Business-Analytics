## *Loading Data*

```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  

data=pd.read_csv('/content/Data_students.csv')
print(data)
```

## *Plotting the distribution of scores*

```python
data.plot(x='Hours', y='Scores', style='o')  
plt.title('Hours vs Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')  
plt.show()
```

## *preparation of data*

```python
from sklearn.model_selection import train_test_split  

X = data.iloc[:, :-1].values  
y = data.iloc[:, 1].values 

X_train, X_test, y_train, y_test = train_test_split(X, y, 
                            test_size=0.2, random_state=0)
```

## *Linear Regression Algorithm*

```python
from sklearn.linear_model import LinearRegression  

LR = LinearRegression()  
LR.fit(X_train, y_train)

line_equation = LR.coef_*X+LR.intercept_

# Plotting for the test data
plt.scatter(X, y)
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.plot(X, line_equation);
plt.show()
```

## *Prediction Scores*

```python
Predicted_y = LR.predict(X_test)

df = pd.DataFrame({'Actual': y_test, 'Predicted': Predicted_y})  
df
```

```python
from sklearn import metrics

my_hours = 9.25
algorithm_pred = LR.predict([[my_hours]])
print("required Hours = ",my_hours)
print("Accuracy = ",algorithm_pred[0])
print('Mean Absolute Error:',metrics.mean_absolute_error(y_test, Predicted_y))
```
