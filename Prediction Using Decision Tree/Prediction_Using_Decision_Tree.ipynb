## *Loading Iris Data*

```python
import pandas as pd

# Loading the iris dataset
iris=pd.read_csv("/content/Iris.csv")

# Making the iris as a dataframe
df=pd.DataFrame(iris)
df=df.drop(["Id"],axis=1)
print(df.head())
# Our target 
y=df['Species']
print(y)
```

## *Decision Tree Algorithm*

```python
import sklearn
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score


dtc=DecisionTreeClassifier()
X=df.drop(['Species'],axis=1)
Y=df['Species']
X_train, X_test, y_train, y_test = train_test_split(X, Y,test_size=0.2,random_state=0)
dtc = dtc.fit(X_train, y_train)
predicted_y = dtc.predict(X_test)
Acc = accuracy_score(y_test, predicted_y)
print("The Accuracy is : ",(Acc *100),"%")
```

## *Plotting The Decision Tree Graph*

```python
from sklearn import tree
from six import StringIO 
import  matplotlib.pyplot as plt

fig = plt.subplots(figsize=(20, 8))

tree.plot_tree(dtc,feature_names=df.iloc[:, :-1].columns,
                           class_names=['0','1','2'],
                           filled=True, rounded=True)
plt.show()
```
